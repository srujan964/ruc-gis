```python
import geopandas as gpd
import pandas as pd
import matplotlib.pyplot as plt

import tarfile
```

```python
# USGS Small Scale States Boundaries of the US shapefile - https://www.sciencebase.gov/catalog/item/581d052de4b08da350d524e5
!wget -q -O us-shapefile.tar.gz https://prd-tnm.s3.amazonaws.com/StagedProducts/Small-scale/data/Boundaries/statesp010g.shp_nt00938.tar.gz
```

```python
with tarfile.open('us-shapefile.tar.gz') as tf:
  tf.extractall(filter='data')
```

```python
states = gpd.read_file('statesp010g.shp')
states.plot()
```

```python
# Default plot has a lot of negative space on the right hand side
# Adjusting the axes and constraining it only to the mapped region
fig, ax = plt.subplots(figsize = (20, 15))
ax.set_xlim([-200, -40])
states.plot(ax=ax)
```
